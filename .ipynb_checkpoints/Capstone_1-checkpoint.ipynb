```python
#scanning each column for wrong data types
def bad_datatype(series, datatype):
    '''bad_datatype(series, datatype)
        Iterates through a pandas Series for elements that do not have the type 'datatype'
        returns name of series and thenumber of faulty values.'''
    count = 0
    for item in series:
        if type(item) != datatype:
            count +=1
    return series.name, count

print( bad_datatype(bids.auction, str))
print( bad_datatype(bids.merchandise, str))
print( bad_datatype(bids.device, str))
print( bad_datatype(bids.time, np.int64))
print( bad_datatype(bids.country, str))
print( bad_datatype(bids.ip, str))
print( bad_datatype(bids.url, str))
print( bad_datatype(train.payment_account, str))
print( bad_datatype(train.address, str))
```

```python
#looking deeper into bids.country to see what the faulty data types are and if it would be important to exclude them.
name, count = bad_datatype(bids.country, str)
double_check = 0
for item in bids.country:
    if type(item) != str:
        print (type(item), item)
        double_check +=1
print (len(bids.country), count/len(bids.country), double_check)
```

```python
bot_avg = []
human_avg = []
count=0
for auction_name in mergefirst.auction.unique():
    bot_avg.append(mergefirst[mergefirst.auction == auction_name][mergefirst.outcome==1].time.mean())
    human_avg.append(mergefirst[mergefirst.auction == auction_name][mergefirst.outcome==0].time.mean())
    print (count)
    count+=1
```

```python
#Plot of average bid time per human/bot
plt.subplot(211)
plt.plot(np.arange(len(human_avg)), human_avg, marker='.', linestyle='None', color='r', alpha=0.1)
plt.plot(np.arange(len(bot_avg)), bot_avg, marker='.', linestyle='None', color='b', alpha=0.1)
plt.subplot(212)
plt.plot(np.arange(len(bot_avg)), bot_avg, marker='.', linestyle='None', color='b', alpha=0.1)
plt.plot(np.arange(len(human_avg)), human_avg, marker='.', linestyle='None', color='r', alpha=0.1)

plt.show()
```
