```python
import requests
import json
import time
import datetime
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, accuracy
from surprise.model_selection import KFold, train_test_split, cross_validate
from surprise import KNNBasic
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from surprise.prediction_algorithms.knns import *
from surprise.prediction_algorithms.slope_one import SlopeOne
from surprise.prediction_algorithms.baseline_only import BaselineOnly
from surprise.prediction_algorithms.matrix_factorization import NMF
from surprise.prediction_algorithms.matrix_factorization import SVD
import matplotlib.pyplot as plt
```

This notebook is used for the code to build and evaluate the recommender model. 

```python
#Loaduserlist from txt and double check its formatting
users = np.genfromtxt('users.txt', delimiter=',').astype(int)
users
```

```python
#Reloading the mastery dataframe from file to double check
reclaim_mastery_df = pd.DataFrame.from_csv("mastery_df.csv")
mFill_df = reclaim_mastery_df.fillna(0)
mFill_df.head()
```

```python
def df2surprise(df):
    '''df2surprise is a function which takes a normal pandas DataFrame and converts it into a form that can be used in surprise.
    Surprise takes in dataframes with columns = user, item, score'''
    data = pd.DataFrame(df.stack()).reset_index()
    data.columns=['sumId', 'ChampId', 'points']
    surprise_data = Dataset.load_from_df(data, reader)
    return surprise_data
```

```python
#Creation of scaled Data sets using the standard scaler and convert to surprise format
scaler = StandardScaler()
#scale Columns
test = scaler.fit_transform(mFill_df)
col_scaled = pd.DataFrame(test, index=mFill_df.index, columns=mFill_df.columns)
#scale Rows
test = scaler.fit_transform(mFill_df.T)
row_scaled = pd.DataFrame(test.T, index=mFill_df.index, columns=mFill_df.columns)
#scale Rows then Columns
test = scaler.fit_transform(mFill_df.T)
test2 = scaler.fit_transform(test.T)
row_col_scaled = pd.DataFrame(test2, index=mFill_df.index, columns=mFill_df.columns)
#Scale Columns then rows
test = scaler.fit_transform(mFill_df)
test2 = scaler.fit_transform(test.T)
col_row_scaled = pd.DataFrame(test2.T, index=mFill_df.index, columns=mFill_df.columns)

reader = Reader()
row_scaled_data = df2surprise(row_scaled)
col_scaled_data = df2surprise(col_scaled)
row_col_scaled_data = df2surprise(row_col_scaled)
col_row_scaled_data = df2surprise(col_row_scaled)
```

```python
#Confirm that scaler is working properly (row based data is using Transpose since normalization was done via row)
for data in [col_scaled, row_scaled.T, row_col_scaled.T, col_row_scaled]:
    print('Mean: {}, Std: {}'.format(data.mean().iloc[0], data.std().iloc[0]))
```

All the data is normalized correctly but the hybrid row and column approaches add error.

```python
def test_model(algo):
    '''function used to test model on all the standard scaler datasets'''
    for data in [col_scaled_data, row_scaled_data, row_col_scaled_data, col_row_scaled_data]:
        t = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
base = BaselineOnly()
test_model(base)
```

```python
slopeone = SlopeOne()
test_model(slopeone)
```

```python
nmf = NMF()
test_model(nmf)
```

```python
svd = SVD()
test_model(svd)
```

```python
#Testing k-nn with different similarity
cos_sim = {'name': 'cosine'}
msd_sim = {'name' : 'msd'}
pears_sim = {'name': 'pearson'}
cos_knn = KNNBasic(sim_options=cos_sim)
msd_knn = KNNBasic(sim_options=msd_sim)
pears_knn = KNNBasic(sim_options=pears_sim)
print("Cosine Similarities\n")
test_model(cos_knn)
'''print("\nMean Squared Distance Similarities\n")
test_model(msd_knn)
print("\nPearson Similarities\n")
test_model(pears_knn)
'''
```

```python
item_cos_sim = {'name': 'cosine', 'user_based':False}
item_msd_sim = {'name' : 'msd', 'user_based':False}
item_pears_sim = {'name': 'pearson', 'user_based':False}
item_cos_knn = KNNBasic(sim_options=item_cos_sim)
item_msd_knn = KNNBasic(sim_options=item_msd_sim)
item_pears_knn = KNNBasic(sim_options=item_pears_sim)
print("Item Based Similarities using K-nn")
print("Cosine Similarities\n")
test_model(item_cos_knn)
print("\nMean Squared Distance Similarities\n")
test_model(item_msd_knn)
print("\nPearson Similarities\n")
test_model(item_pears_knn)
```

After viewing a few of the predictions we see that there are no predictions less than 1 even though all the results are close to zero when the data was scaled using a standard scaler. This must be why most of the RMSE scores are similar when using that data. After seeing a minimum prediction score I am concerned that there is a maximum one as well so I tested a few values and you can see the final result. The baseline model is not predicting above a score of 5 even though the real scores have been scalted to 10 at minimum. Moving forward with this we will scale the data between 1-5 to account for this range.

```python
#create actual min max scaled data for the appropriate range.
minmax = MinMaxScaler(feature_range=(1,5))
test = minmax.fit_transform(mFill_df)
minmax_scaled = pd.DataFrame(test, index=mFill_df.index, columns=mFill_df.columns)
minmax_scaled = df2surprise(minmax_scaled)
test = minmax.fit_transform(mFill_df.T)
rowminmax_scaled = pd.DataFrame(test.T, index=mFill_df.index, columns=mFill_df.columns)
rowminmax_scaled = df2surprise(rowminmax_scaled)

base = BaselineOnly()
slopeone = SlopeOne()
svd = SVD()
nmf = NMF()
```

```python
#Double check that the min max scaler is working as intended (i.e. avoiding prediction limitations)
train, test = train_test_split(rowminmax_scaled, test_size=0.0001, random_state=42)
base.fit(train)
pred = base.test(test, verbose=True)
```
