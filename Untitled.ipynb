```python
#scanning each column for wrong data types
def bad_datatype(series, datatype):
    '''bad_datatype(series, datatype)
        Iterates through a pandas Series for elements that do not have the type 'datatype'
        returns name of series and thenumber of faulty values.'''
    count = 0
    for item in series:
        if type(item) != datatype:
            count +=1
    return series.name, count

print( bad_datatype(bids.auction, str))
print( bad_datatype(bids.merchandise, str))
print( bad_datatype(bids.device, str))
print( bad_datatype(bids.time, np.int64))
print( bad_datatype(bids.country, str))
print( bad_datatype(bids.ip, str))
print( bad_datatype(bids.url, str))
print( bad_datatype(train.payment_account, str))
print( bad_datatype(train.address, str))
```
